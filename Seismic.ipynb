## USING A BETTER MODEL ON A DIFFRENT METHOD OF CLASS BALANCING

Building a new model (ANN) in tensorflow/keras for the balanced class

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

def focal_loss(gamma=2., alpha=0.25):
    def focal_loss_fixed(y_true, y_pred):
        y_true = tf.convert_to_tensor(y_true, tf.float32)
        y_pred = tf.convert_to_tensor(y_pred, tf.float32)

        alpha_t = y_true * alpha + (1 - y_true) * (1 - alpha)
        p_t = y_true * y_pred + (1 - y_true) * (1 - y_pred)
        fl = - alpha_t * tf.pow((1 - p_t), gamma) * tf.math.log(p_t)
        return tf.reduce_mean(fl)
    return focal_loss_fixed

def ANN(X_train, y_train, X_test, y_test, loss, weights):
    # Adjust the input_dim to match the number of features in your data
    input_dim = X_train.shape[1]

    model = models.Sequential([
        layers.Dense(input_dim, input_dim=input_dim, activation='relu'),
        layers.Dense(15, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam', loss=loss, metrics=['accuracy'])

    if weights == -1:
        model.fit(X_train, y_train, epochs=10)
    else:
        model.fit(X_train, y_train, epochs=10, class_weight=weights)

    evaluation = model.evaluate(X_test, y_test)
    print(f"Model evaluation: {evaluation}")

    y_preds = model.predict(X_test)
    y_preds = np.round(y_preds).astype(int)

    print("Classification Report without class balancing: \n", classification_report(y_test, y_preds))

    return y_preds

```

```python
# Define your loss function as a string (e.g., 'binary_crossentropy')
loss_function = 'binary_crossentropy'

# Call the ANN function with the specified loss function and weights
y_preds = ANN(X_train, y_train, X_test, y_test, loss_function, -1)
```

# Method 1: Undersampling

```python
# Class count
count_class_0, count_class_1 = data.Class.value_counts()

# Divide by class
df_class_0 = data[data['Class'] == 0]
df_class_1 = data[data['Class'] == 1]
```

```python
# Undersample 0-class and concat the DataFrames of both class
df_class_0_under = df_class_0.sample(count_class_1)
df_test_under = pd.concat([df_class_0_under, df_class_1], axis=0)

print('Random under-sampling:')
print(df_test_under.Class.value_counts())
```

```python
y = df_test_under['Class'].astype(np.float32)
X = df_test_under[['V18', 'V17', 'V16','V14','V12','V11','V10','V4','V3','V1']]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)
```

```python
# Number of classes in training Data
y_train.value_counts()
```

```python
y_preds = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)

from sklearn.metrics import precision_recall_curve, auc


# Calculate precision-recall pairs for different probability thresholds
precision, recall, _ = precision_recall_curve(y_test, y_preds)

# Compute AUPRC
auprc = auc(recall, precision)

print(f"AUPRC: {auprc}")
```

## Method2: Oversampling

```python
# Oversample 1-class and concat the DataFrames of both classes
df_class_1_over = df_class_1.sample(count_class_0, replace=True)
df_test_over = pd.concat([df_class_0, df_class_1_over], axis=0)

print('Random over-sampling:')
print(df_test_over.Class.value_counts())
```

```python
y = df_test_over['Class'].astype(np.float32)
X = df_test_over[['V18', 'V17', 'V16','V14','V12','V11','V10','V4','V3','V1']]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)
```

```python
# Number of classes in training Data
y_train.value_counts()
```

```python
y_preds = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)

from sklearn.metrics import precision_recall_curve, auc


# Calculate precision-recall pairs for different probability thresholds
precision, recall, _ = precision_recall_curve(y_test, y_preds)

# Compute AUPRC
auprc = auc(recall, precision)

print(f"AUPRC: {auprc}")
```
